```python
random.shuffle(data)
```

```python
train = []
validate = []
test = []

def data_preparation(train_size, validate_size, test_size):
  for s in data:
    num = random.random()
    if num >= 0 and num <= train_size:
      train.append(s)
    elif num >= 0 and num <= train_size + validate_size:
      validate.append(s)
    else:
      test.append(s)
data_preparation(.7,.2,.1)
```

# Classification

```python
import math
```

```python
def calculate_euclidean_distance(validation, train):
  euclidean_distance = 0
  for x, y in zip(validation, train):
    euclidean_distance = euclidean_distance + (x-y)**2
  return math.sqrt(euclidean_distance)
```

```python
def sortfunc(distance):
  return distance[1]
```

```python
def kNN_classifier(k, train, val):
  distance = []
  accuracy = 0
  for v in val:
    for t in train:
      euclidean_distance = calculate_euclidean_distance(v[:-1],t[:-1])
      distance.append([t,euclidean_distance])
    distance.sort(key=sortfunc)
    select = distance[:3]
    label = {
        0 : 0,
        1 : 0,
        2 : 0
    }
    for s in select:
      label[s[0][-1]] += 1
    predicted_class = max (label, key=label.get)
    if v[-1] - predicted_class == 0:
      accuracy += 1
  return accuracy/ len(val)
```

```python
print(kNN_classifier(1, train, validate))
```
